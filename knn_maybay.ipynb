```python
data = pd.read_csv('/content/drive/MyDrive/Code_co_quynh_anh/train_df.csv')
# data=pd.read_csv('preprocessed_train_data.csv')
print(data.shape)
```

```python
data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
data['experiment'] = LE.fit_transform(data['experiment'])
```

```python
X = data.drop('event', axis=1)
```

```python
X.head(10)
```

```python
y = data['event'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state = 0)
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors = 1, p = 2)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test[0:200])

print("Print results for 20 test data points:")
print("Predicted labels: ", y_pred[0:200])
print("Ground truth    : ", y_test[0:200])
```

```python
count = 0
for m,n in zip(y_pred[0:200],y_test[0:200]):
  if m == n:
    count=count+1
print(f'tỉ lệ Predicted đúng là {count}/{len(y_pred[0:200])}')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test[0:200],y_pred[0:200]))
```
